```python
sales =float(input("Enter the Sales Amount: $"))
taxrate = float(input("Enter the Tax Rate (%): "))

cogs = .4 * sales
grossprofit = sales - cogs
tax = (taxrate/100) * grossprofit
netincome = grossprofit - tax

print(f"Sales: ${sales}")
print(f"Cost of goods sold: ${cogs}")
print(f"Gross profit: ${grossprofit}")
print(f"Tax: ${tax}")
print(f"Net income: ${netincome}")

```

```python
n1 = int(input("Enter the first number: "))
opp = input("Enter the operation (+, -, *, /): ")
n2 = int(input("Enter the second number: "))


if opp == "+":
  print(n1 + n2)
elif opp == "-":
  print(n1 - n2)
elif opp == "*":
  print(n1 * n2)
elif opp == "/":
  print(n1 / n2)
else:
  print("Invalid operation")
```

```python
flatfee = 5
varfee = 2.5

for hours in range(1,9):
  cost = flatfee + (varfee*hours)
  if cost < 10:
    cost = 10.0
  elif cost > 20:
    cost = 20.0
  print(f"Hour {hours}: ${cost}")
```

```python
flatfee = 5
varfee = 2.5
hours = 1

while hours < 9:
  cost = flatfee + (varfee*hours)
  if cost < 10:
    cost = 10.0
  elif cost > 20:
    cost = 20.0
  print(f"Hour {hours}: ${cost}")
  hours += 1
```

```python
fixedcost = 200
ticprice = 10
att = 20

print("$Ad\tProfit")
for adcost in range(0,225,25):
  addatt = 2 * ((adcost)**.5)
  totalcost = fixedcost + adcost
  gross = (ticprice * att) + (round(addatt) * ticprice)
  net = gross - totalcost
  print(f" {float(adcost)}\t{float(net)}")
```

```python
fixedcost = 200
ticprice = 10
att = 20
adcost = 0

print("$Ad\tProfit")
while adcost < 225:
  adcost += 25
  addatt = 2 * ((adcost)**.5)
  totalcost = fixedcost + adcost
  gross = (ticprice * att) + (round(addatt) * ticprice)
  net = gross - totalcost
  print(f" {float(adcost)}\t{float(net)}")
```
