```python
def roi(cost,revenue):
  assert isinstance(cost, (int, float)), "Cost must be a number"
  assert isinstance(revenue,(int, float)), "Revenue must be a number"
  assert cost != 0, "Cost must be greater than 0"
  roi = float((revenue - cost)/(cost))
  print(f"ROI: {roi}")

roi(50,'abc')
```

```python
def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float)), "Principal must be a number"
  assert isinstance(annual_rate, (int, float)), "annual rate must be a number"
  assert isinstance(months, (int, float)), "months must be a number"

  if (principal <=0) or (annual_rate <= 0) or (months <= 0):
    raise ValueError("Invalid input")

  try:
    r = annual_rate/12
    if r != 0:
      pmt = (r* principal) / (1 - (1+r)**-months)
      print(pmt)
    elif r == 0:
      pmt = principal/months
      print(pmt)
  except Exception as e:
    print(e)

loan_payment(1000, .10, 12)
```

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def display(self, formal=False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(self.first)

  def __str__(self):
    return "Instance of Class 'Greeting'"

```

```python
from typing import Self
class point:
  def __init__(self, x, y, x2, y2):
    self.x = x
    self.y = y
    self.x2 = x2
    self.y2 = y2

  def distance(self):
    self.distance = ((self.x - self.x2)**2 + (self.y - self.y2)**2)**.5
    return self.distance

class rectangle:
  def __init__(self, x1, y1, x2, y2, x3, y3, x4, y4 ):
    self.x1 = x1
    self.y1 = y1
    self.x2 = x2
    self.y2 = y2
    self.x3 = x3
    self.y3 = y3
    self.x4 = x4
    self.y4 = y4

  def width(self):
    self.width = ((self.x2 - self.x1)**2 + (self.y2 - self.y1)**2)**.5
    return self.width

  def height(self):
    self.height = ((self.x3 - self.x2)**2 + (self.y3 - self.y2)**2)**.5

  def area(self):
    self.area = self.width * self.height
    return self.area


d = point(0,0,3,4)
print(f"Distance between the 2 points: {d.distance()}")

r = rectangle(0,0, 0,3 , 3,3, 3,0)
print(r.width())
print(r.height())
print(r.area())
```

```python
from logging import Manager
class Employee:
  def __init__(self, name, wage, hours):
    self.name = name
    self.__wage = wage
    self.hours = hours

  def calc_pay(self):
    return self.__wage * self.hours

  def __str__(self):
    return f"Employee: {self.name}"

  def get_wage(self):
    return self.__wage

  def set_wage(self, newWage):
    assert newWage > 0, "Wage must be greater than 0"
    self.__wage = newWage

class Manager(Employee):
  def __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus

  def calc_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, .2)

print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly pay: {employee.calc_pay()}")
print(f"Manager {manager.name} weekly pay: {manager.calc_pay()} that includes 20% bonus")
```
